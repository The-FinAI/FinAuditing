```python
import json
import pandas as pd
import numpy as np
from tqdm.std import tqdm
import re
from typing import List
import codecs
import ast
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import classification_report
```

```python
result_path = ""
```

```python
data = []
with open(result_path, "r") as f:
    for line in tqdm(f):
        json_line = json.loads(line)
        data.append(json_line)
```

```python
# data[2]
```

```python
##### for the results generated from lm-evaluation-harness framework

def parse_result(json_data):
    true_answer = []
    pred_answer = []
    
    for json_line in tqdm(json_data): 
        target = ast.literal_eval(json_line.get("doc")["answer"])
        resps = json_line.get("filtered_resps")[0]
        
        if resps == '[invalid]':
            resps = 'invalid'
        true_answer.append(target)
        pred_answer.append(resps)

    return true_answer, pred_answer
```

```python
true_answer, pred_answer = parse_result(data)
```

```python
## 'invalid'
print(classification_report(true_answer, pred_answer, labels=['Reversal', 'Inappropriateness', 'CombinationErr'], digits=4))
```
